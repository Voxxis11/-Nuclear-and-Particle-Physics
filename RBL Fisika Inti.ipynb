```python
import matplotlib.pyplot as plt
import numpy as np
import pynucastro as pyna

rl = pyna.ReacLibLibrary()
c12pg = rl.get_rate_by_name("c12(p,g)n13")
n15pa = rl.get_rate_by_name("n15(p,a)c12")

# Define temperature range
temperature_range = np.linspace(1e8, 5e9, 100)

# Calculate rates for c12pg and n15pa reactions
c12pg_rates = [c12pg.eval(t) for t in temperature_range]
n15pa_rates = [n15pa.eval(t) for t in temperature_range]

# Plot the data for both reactions
fig, ax = plt.subplots()

ax.plot(temperature_range, c12pg_rates, label="c12(p,g)n13")
ax.plot(temperature_range, n15pa_rates, label="n15(p,a)c12")

# Set labels and title
ax.set_xlabel('Temperature (K)')
ax.set_ylabel('Rate (cm^3/mol/s)')
ax.set_title('Comparison of n15(p,a)c12 and c12(p,g)n13 Rates')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pynucastro as pyna

rl = pyna.ReacLibLibrary()
c12pg = rl.get_rate_by_name("c12(p,g)n13")
n15pa = rl.get_rate_by_name("n15(p,a)c12")

# Extract temperature and rate data for c12pg
c12pg_temps = c12pg.get_temperatures()
c12pg_rates = c12pg.get_rates()

# Extract temperature and rate data for n15pa
n15pa_temps = n15pa.get_temperatures()
n15pa_rates = n15pa.get_rates()

# Plot the data for both reactions
fig, ax = plt.subplots()

ax.plot(c12pg_temps, c12pg_rates, label="c12(p,g)n13")
ax.plot(n15pa_temps, n15pa_rates, label="n15(p,a)c12")

# Set labels and title
ax.set_xlabel('Temperature (K)')
ax.set_ylabel('Rate (cm^3/mol/s)')
ax.set_title('Comparison of n15(p,a)c12 and c12(p,g)n13 Rates')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
